```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis
```

```python
df=pd.read_csv("salary.csv") # Read the CSV file into a DataFrame
```

```python
df.head() # Display the first five rows of the DataFrame
```

```python
df = df.drop('salary', axis=1) # Drop the 'salary' column from the DataFrame
```

```python
df.head()
```

```python
df.drop('salary_currency',axis=1) # Drop the 'salary_currency' column from the DataFrame
```

```python
df_new = df.dropna() # Drop rows with missing values
```

```python
df_new
```

```python
duplicate_data = df_new[df_new.duplicated()]  # Find duplicate rows in the DataFrame
```

```python
df_new.duplicated()
```

```python
df_unique=df.drop_duplicates()# dropping the duplicate rows
```

```python
# getting unique values for all column in df
unique_values= {column: df_unique[column].unique() for column in df_unique.columns} 
```

```python
#printing in organised manner
for column, values in unique_values.items():
    print('Unique values in', column, ":")
    print(values)
    print()
```

```python
#Renaming the columns
df['experience_level'] = df['experience_level'].replace({'SE': 'Senior Expert',
                                                        'MI': 'Medium Intermediate',
                                                        'EN': 'Entry Level',
                                                        'EX': 'Executive Level'})
```

```python
df['experience_level'] 
```

```python
# Calculating summary statistics,skewness and kurtosis for the 'salary_in_usd' column
summary_stats = df['salary_in_usd'].describe()
skewness = skew(df['salary_in_usd'])
kurt = kurtosis(df['salary_in_usd'])
print(summary_stats)
print('skew: ',skewness)
print('kurt: ',kurt)
```

```python
correlation_matrix = df.corr(numeric_only=True) # Calculating correlation matrix of all numeric columns in df
correlation_matrix
```

```python
# using groupby function in df's salary_in_usd by 'experience_level' and calculating the mean salary 
salaries_by_experience = df.groupby('experience_level')['salary_in_usd'].mean()
# Plotting the chart
salaries_by_experience.plot(kind='bar')
# Adding labels and title to the plot
plt.xlabel('Experience Level')
plt.ylabel('Mean Salary')
plt.title('Mean Salary by Experience Level')
plt.show()
print(salaries_by_experience)
```

```python
#using .plot function to plot a chart for a column in dataframe
df['remote_ratio'].plot(kind='hist', color='green')
# Adding labels and title to the plot
plt.xlabel('Remote Ratio', fontsize=20)
plt.ylabel('Frequency', fontsize=20)
plt.title('Histogram of Remote Ratio', fontsize=24)
plt.grid()
plt.show()
```

```python
#boxplotting remote_ratio column
df.boxplot(column='remote_ratio')
plt.ylabel('Remote Ratio', fontsize=20)
plt.title('Box Plot of Remote Ratio', fontsize=20)
plt.show()
```
